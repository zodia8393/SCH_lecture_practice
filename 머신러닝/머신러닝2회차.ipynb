```python
import pandas as pd
import numpy as np
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'],
'age': [42, np.nan, 36, 24, 73],
'sex': ['m', np.nan, 'f', 'm', 'f'],
'preTestScore': [4, np.nan, np.nan, 2, 3],
'postTestScore': [25, np.nan, np.nan, 62, 70]}
df=pd.DataFrame(raw_data,columns=['first_name','last_name','age','sex','preTestScore','postTestScore'])
df
```

```python
df.isnull().sum()/len(df) 
#결측치 확인시 isnull함수 ,NaN값 존재시 True 아니면 False
#sum함수로 True인경우 모두 더하고 전체 데이터 수로 나눠 열별 데이터 결측치 비율을 구한다
```

```python
df.dropna()
#드롭 : 결측치가 나온 열이나 행 삭제
#dropna 사용하여 NaN이 있는 모든 데이터 행을 없앤다
```

```python
#드롭 결과물 저장
df_no_missing=df.dropna()
df_no_missing
```

```python
df_cleaned=df.dropna(how='all')
df_cleaned
```

```python
df['location']=np.nan
df.dropna(axis=1,how='all')
```

```python
df.dropna(axis=0,thresh=1) 
#thresh 는 데이터의 개수를 기준으로 하는 매개변수
```

```python
df.dropna(thresh=5) #데이터가 5개 이상있으면 데이터를 남긴다
```

```python
df.fillna(0)
```

```python
df['preTestScore'].fillna(df['preTestScore'].mean(),inplace=True)
df
#열 단위의 평균값을 계산하여 해당 열에만 값을 채운다
```

```python
df.groupby('sex')[
    "postTestScore"].transform('mean') 
#groupby 함수로 각 인덱스의  성별에 따라 빈칸을 채운다
```

```python
df['postTestScore'].fillna(
    df.groupby('sex')['postTestScore'].transform('mean'),
inplace=True)
df
```

```python
weight_dict = {3:"M", 4:"L", 5:"XL"}
edges["weight_sign"] = edges["weight"].map(weight_dict)
weight_sign = pd.get_dummies(edges["weight_sign"])
weight_sign
```

```python
pd.concat([edges,weight_sign],axis=1)
#데이터를 원핫인코딩 형태로 변경후 필요에 따라 병합이나 연결로 두가지의 데이터를 합친다
```

```python
#바인딩 : 연속형데이터 -> 범주형 데이터 변환
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks',
'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts',
'Scouts', 'Scouts', 'Scouts'],
'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st',
'1st', '2nd', '2nd'],
'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon',
'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}

df=pd.DataFrame(raw_data,columns=['regiment','company','name','preTestScore','postTestScore'])
df
```

```python
bins=[0,25,50,75,100]
group_names=['Low','Okay','Good','Great']
categories=pd.cut(
    df['postTestScore'],bins,labels=group_names)
categories 

#cut함수로 나눌 시리즈 객체와 구간,구간의 이름을 넣어주면 해당 값을 바인딩해서 표기함
```
